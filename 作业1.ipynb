```python
# 最小二乘法
def least_squares(x, y):
    x_mean = np.mean(x)
    y_mean = np.mean(y)
    a = np.sum((x - x_mean) * (y - y_mean)) / np.sum((x - x_mean) ** 2)
    b = y_mean - a * x_mean
    return a, b

# 梯度下降法
def gradient_descent(x, y, lr=0.01, epoch=1000):
    a = 0
    b = 0
    for _ in range(epoch):
        a_grad = -2 * np.mean(x * (y - (a * x + b)))
        b_grad = -2 * np.mean(y - (a * x + b))
        a -= lr * a_grad
        b -= lr * b_grad
    return a, b

# 牛顿法
def newton(x, y, max_iter=1000):
    # 初始化参数
    a, b = 0.0, 0.0
    theta = np.array([a, b])
    
    for i in range(max_iter):
        # 计算梯度
        grad_a = -2 * np.sum(x * (y - (theta[0] * x + theta[1])))
        grad_b = -2 * np.sum(y - (theta[0] * x + theta[1]))
        gradient = np.array([grad_a, grad_b])
        
        # 计算 Hessian 矩阵
        H = np.array([
            [2 * np.sum(x**2), 2 * np.sum(x)],
            [2 * np.sum(x), 2 * len(x)]
        ])
        
        # 更新参数
        delta = np.linalg.inv(H) @ gradient
        theta_new = theta - delta
          
        theta = theta_new
    
    return theta[0], theta[1]



# 计算误差
def compute_error(y_pred, y):
    return np.mean((y_pred - y) ** 2)

# 训练和测试
a_ls, b_ls = least_squares(x_train, y_train)
a_gd, b_gd = gradient_descent(x_train, y_train)
a_nt, b_nt = newton(x_train, y_train)

# 参数输出
print("Least Squares:")
print("a:", a_ls, "b:", b_ls)
print("Gradient Descent:")
print("a:", a_gd, "b:", b_gd)
print("Newton's Method:")
print("a:", a_nt, "b:", b_nt)

# 训练误差
y_pred_ls = a_ls * x_train + b_ls
y_pred_gd = a_gd * x_train + b_gd
y_pred_nt = a_nt * x_train + b_nt

error_ls = compute_error(y_pred_ls, y_train)
error_gd = compute_error(y_pred_gd, y_train)
error_nt = compute_error(y_pred_nt, y_train)

print("训练误差")
print("Least Squares Error:", error_ls)
print("Gradient Descent Error:", error_gd)
print("Newton's Method Error:", error_nt)

# 测试误差
y_pred_ls = a_ls * x_test + b_ls
y_pred_gd = a_gd * x_test + b_gd
y_pred_nt = a_nt * x_test + b_nt

error_ls = compute_error(y_pred_ls, y_test)
error_gd = compute_error(y_pred_gd, y_test)
error_nt = compute_error(y_pred_nt, y_test)

print("测试误差")
print("Least Squares Error:", error_ls)
print("Gradient Descent Error:", error_gd)
print("Newton's Method Error:", error_nt)
```

```python
# 非线性拟合
from lmfit import Model

# 正弦型函数
def model_fuc(x, a, b, c, d):
    return a * np.sin(b * x + c) + d

# 从数据中估计初始参数
a_guess = (np.max(y_train) - np.min(y_train)) / 2  # 估计振幅
d_guess = np.mean(y_train)  # 估计偏移量
fft_values = np.fft.fft(y_train - d_guess)  # 去偏移后计算 FFT
frequencies = np.fft.fftfreq(len(x_train), x_train[1] - x_train[0])
b_guess = np.abs(frequencies[np.argmax(np.abs(fft_values))]) * 2 * np.pi  # 估计频率
c_guess = 0  # 初始相位设为 0

# 创建模型
model = Model(model_fuc)
# 定义参数
params = model.make_params(a=a_guess, b=b_guess, c=c_guess, d=d_guess)
# params = model.make_params(a=1, b=3, c=0, d=0)
params['c'].set(min=0, max=2*np.pi)  # c 的范围 [0, 2π]
params['d'].set(min=-np.inf, max=np.inf)  # d 无限制

# 拟合
result = model.fit(y_train, params, x=x_train, method='leastsq')

# 输出结果
print(result.fit_report())

# 计算误差
y_pred = result.best_fit
error = compute_error(y_pred, y_train)
print("训练误差:", error)
error = compute_error(y_pred, y_test)
print("测试误差:", error)

# 绘制结果
plt.figure(figsize=(10, 6))
plt.scatter(x_train, y_train, color='blue', label='train data')
plt.plot(x_train, result.best_fit, color='red', label='fitting curve')
plt.title('fitting curve')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
